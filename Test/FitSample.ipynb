## Loop fit

```python
i = 0
slopes = []
while i < len(y) - 12:
    x_temp = x
    y_temp = y[i:i+12]
    fit = np.polyfit(x_temp, y_temp, 1)
    slope_temp = fit[0]
    slopes.append(slope_temp)
    i += 1
##format converting
slopes = np.array(slopes)
```
